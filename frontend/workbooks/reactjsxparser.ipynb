```typescript
import { Parser } from "acorn";
import acornJsx from "acorn-jsx";

const acornParser = Parser.extend(
  acornJsx(),
);
const acornOptions = {
    ecmaVersion: "latest" as const, 
   // sourceType: "module", 
   plugins: {jsx: true}, // unsure if this is needed
   // location: false,
   // allowReturnOutsideFunction: true,
}

// pass in a string of code and get back an AST
function acornParse(codeString) {
    return acornParser.parse(codeString, acornOptions);
} 
```

```typescript
const babel = require('@babel/standalone');

// Your JSX function string
const code = `
  function MyComponent() {
    return (
      <div>
        {/* This is a comment */}
        <h1>Hello, world!</h1>
        {/* Another comment */}
        <p>Welcome to my component</p>
      </div>
    );
  }
`;

// Parse the code using Babel
const ast = babel.parse(code, {
  plugins: ['jsx'],
});

// Function to remove comments from JSX nodes
function removeComments(node) {
  if (node.type === 'JSXElement') {
    node.openingElement.attributes = node.openingElement.attributes.filter(attr => attr.type !== 'JSXSpreadAttribute');
    node.children = node.children.filter(child => !(child.type === 'JSXExpressionContainer' && child.expression.type === 'JSXEmptyExpression'));
  }
  return node;
}

// Traverse the AST to find the XML chunk inside the return statement
let xmlChunk = '';
babel.traverse(ast, {
  FunctionDeclaration(path) {
    path.traverse({
      ReturnStatement(returnPath) {
        const argument = returnPath.get('argument');
        if (argument.isJSXElement()) {
          xmlChunk = babel.transformFromAstSync(removeComments(argument.node), null, {
            configFile: false,
            plugins: ['@babel/plugin-transform-react-jsx'],
          }).code;
        }
      },
    });
  },
});

// Print the XML chunk without comments
console.log(xmlChunk);
```
